```python
import csv

with open("web_clients_correct-старое.csv", "r", encoding="utf-8", newline="") as fileread:
    with open("result.txt", "w", encoding="utf-8", newline="") as filewrite:
        csvreader = csv.reader(fileread)
        header = next(csvreader)

        for row in csvreader:
            name = row[0]
            device = row[1]
            browser = row[2]
            sex = row[3]
            age = row[4]
            total = row[5]
            region_of_purchase = row[6]

            device_text = "мобильного" if device == "mobile" or device == "tablet" else "десктопного"
            sex_text = "женского пола" if sex == "female" else "мужского пола"
            verb = "совершила" if sex == "female" else "совершил"

            line = (
                f"Пользователь {name} {sex_text}, {age} лет {verb} покупку на {total} у.е. "
                f"с {device_text} браузера {browser}. "
                f"Регион, из которого совершалась покупка: {region_of_purchase}."
            )

            filewrite.write(line + "\n\n")
```
